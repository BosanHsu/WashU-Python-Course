### Step 1: Estimate the Maximized Total Value and Accumulated Weight of One Warehouse

Read the dataset from  "Products.csv" first.

```python
file_name='Products.csv'
# Write your codes here
f_product = open(file_name,'r')  
product_lines = f_product.readlines()
dic_warehouse = {}
count = 0
for i in range(0, len(product_lines), 3):
    dic_warehouse[count] = []   # Here, the variable "count" denotes the number specified to the warehouse. As the explanation file describes, there are 0~299 warehouses, 300 in total.
    product_weight = product_lines[i].strip('\n').split(',') # Here, we denote the product weight information in this product_weight list. We now use the next line to transform the string data into float data.
    product_weight = [float(i) for i in product_weight]
    
    product_value = product_lines[i + 1].strip('\n').split(',') # The operation on product_value is similar to product_weight.
    product_value = [float(i) for i in product_value]
    for j in range(0, len(product_value)):
        dic_warehouse[count].append([product_weight[j], product_value[j]]) # Here, we record the weight and the value of each product in a list of length 2, 50 of them in total to record the information of 50 products in each warehouse.
    count += 1

# After this code, all the information we need about 300 warehouses and 50 products in each warehouse and the product weight and product value is recorded in the dictionary called "dic_warehouse"

dic_warehouse

```

Now let's get the estimated value and weight for one warehouse.You need to write a function "get_max_value" which could do the following 4 things:
1. Calculate the Value_Weight Ratio for each product. 
2. Sort the calculated Value_Weight Ratios in descending order. 
3. Select the products from the top of the descending order to the bottom while looking at the accumulated         weight of products to make sure the accumulated weight does not exceed the warehouse weight capacity.
4. Calculate the  corresponding estimated value for this warehouse after finishing product selection.

Remember, take every product into consideration; otherwise, you may have the risk of missing products. You want to put as many product with highest Value_Weight Ratio as possible before reaching the weight capacity for this warehouse.

```python
def get_max_value(weight_value_info):          # You should add input variables yourself in the parentheses
    
    # Write your codes here
    weight = []
    value_weight_ratio = []
    value = []
    capacity = 0 # The actual storeage weight of the products in the warehouse, before the operation, we initialize it to zero
    max_capacity = 850 # According to the explanation file, each warehouse should store the product weight not exceeding 850
    value_warehouse = 0 # Similar to capacity, we initialize the value of stored products in the warehouse to zero.
    for product in weight_value_info:
        weight.append(product[0])
        value_weight_ratio.append(product[1] / product[0])
        value.append(product[1])
        
    weight = np.array(weight)
    value_weight_ratio = np.array(value_weight_ratio)
    value = np.array(value
                        )
    indices = np.argsort(- value_weight_ratio)
    value_weight_ratio = value_weight_ratio[indices]
    weight = weight[indices]
    value = value[indices]
    
    for i in range(0, len(weight)):
        if capacity + weight[i] > max_capacity:
            continue # If the capacity exceeds max_capacity(850 in this case, we will have to check if the product next to it to see if its weight is less and can fit in)
        else:
            capacity += weight[i]
            value_warehouse += value[i] 
    
    return(capacity, value_warehouse)           
# Make sure you have calculated the total Value and accumulated weight for one warehouse
get_max_value(dic_warehouse[0])
```

### Step 1: Generate the 300*300 distance matrix and write the results to "Distances.csv"
Let's generate a distance matrix among the 300 warehouses first. Each of the distances can be generated by using a normal distribution with a mean of 600 and a standard deviation of 400. 

Be careful, all the distances generated should be positive numbers and should be rounded to integers using the round() function. After successfully generating the distance matrix, please write it to a new CSV file called “Distances.csv”.

```python
import numpy as np

#create a matrix first(not perfect)
distances = np.abs( np.random.normal( 600, 400, size = [300,300] ).astype(int) ) #abs = abosulte value

#adjust the matrix to a distance matrix
for x in range(300):
    for y in range(300):
        if x == y:
            distances[x][y] = 0  # the distance between a location and itself is 0
        elif x < y:
            distances[x][y] = distances[y][x] # make the distance between a,b =  the distance between b,a(reamin indexs in the right)

distances

# i have used 10*10 matrix for testing, the index are all postive intergers, and the matrix is symmetric
```

Now we are going to save the file in 'Distance.csv'

```python
f = open("Distances.csv", "w") # We will create this file.

for row in distances:
    for col in row:
        f.write( str( col ) + "," )
    f.write("\n")
    
f.close()
```
